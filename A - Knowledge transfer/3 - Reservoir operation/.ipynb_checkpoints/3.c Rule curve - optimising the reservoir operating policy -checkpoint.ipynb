```python
q = np.array(np.random.normal(loc=40,scale=10,size=100),ndmin = 2).transpose()
s_0 = 100
```

```python
def sim_test(q, s_0, R_ref, S_ref, S_max, x):
    # Initialize variables
    T = q.shape[0]
    q_sim = np.array(np.append(np.nan, q), ndmin = 2).transpose()
    u = np.nan*np.zeros((T+1,1))
    r = np.nan*np.zeros((T+1,1))
    s = np.nan*np.zeros((T+1,1))
    v = np.nan*np.zeros((T+1,1))
    V = np.nan*np.zeros((T+1,1))
    s[0] = s_0
    # Run simulation
    for t in np.arange(T):
        # compute the min and max releasable volumes:
        v[t+1] = np.max([s[t] - 100,0])
        V[t+1] = s[t]
        # Compute decision
        u[t] = release_rule(R_ref, S_ref, S_max, s[t],x)
        # Compute release
        r[t+1] = np.max([v[t+1],np.min([V[t+1],u[t]])])
        # Compute future storage
        s[t+1] = s[t] + (q_sim[t+1] - r[t+1])

    # Set parameter values
    w_irr = 50 # water demand m3/d
    S = 1 # water surface area m2
    h_flo = 50 # m
    # Compute step-costs
    h = s / S # water volume - height relationship
    g_flo = np.maximum(h[1:] - h_flo, 0) # m
    g_irr = np.maximum(w_irr - r[1:], 0) # m3/d
    g_flo = np.array(np.append(np.nan, g_flo),ndmin = 2).transpose()
    g_irr = np.array(np.append(np.nan, g_irr),ndmin = 2).transpose()
    
    return s, u, r, v, V, g_flo, g_irr
```

```python
[ s_A, u_A, r_A, v_A, V_A, g_flo_A, g_irr_A ] = sim_test( q, s_0, R_ref, S_ref, S_max, x_A )
```

```python
# Optimizer
from platypus import NSGAII, Problem, Real, Integer

def auto_optim(vars):
    R_ref = vars[0]
    x0 = vars[1]
    x1 = vars[2]
    x2 = vars[3]

    x = [x0, x1, x2]
    s, u, r, v, V, g_flo, g_irr = sim_test(q, s_0, R_ref, S_ref, S_max, x)
    
    J_flo = np.mean(g_flo[1:])
    J_irr = np.mean(g_irr[1:])
    
    constraints = [x1-R_ref]
    
    return [J_flo, J_irr], constraints

problem = Problem(4,2,1)
Real0 = Real(0,100); Real1 = Real(0, np.pi/4);Real2 = Real(0, 200);Real3 = Real(0, np.pi/4)

problem.types[:] = [Real0] + [Real1] + [Real2] + [Real3]
problem.constraints[:] = ">=0"
problem.function = auto_optim

population_size = 30
algorithm = NSGAII(problem,population_size)
algorithm.run(1000) # Number of iterations

results1_optim = np.array([algorithm.result[i].objectives[0] for i in range(population_size)])
results2_optim = np.array([algorithm.result[i].objectives[1] for i in range(population_size)])

solutions_optim_R_ref = [algorithm.result[i].variables[0] for i in range(population_size)]
solutions_optim_x = [algorithm.result[i].variables[1:4] for i in range(population_size)]
```
